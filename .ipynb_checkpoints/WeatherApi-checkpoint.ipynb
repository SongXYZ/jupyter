```python
import json
import requests
```

loads => json=>python对象<br>
dumps => python对象=>json

load => 打开json文件<br>
dump => 写入json文件

```python
weather_file = open("darksky.json", encoding="utf-8")
weather_data = json.load(weather_file)
print(weather_data)
```

```python
url = "https://api.darksky.net/forecast/c6c5eba2e387b1cb7b70b3786d94c0e7/37.8267,-122.4233"
```

```python
reponse = requests.get(url)
print(reponse.text)
```

```python
dark_data = json.loads(reponse.text)
print(dark_data)
```
